```python
ROWS_COUNT = -1
with open("./data/1m_6ex_karel/train.json", "r") as f:
	ROWS_COUNT = sum(1 for _ in f)

print(ROWS_COUNT)
```

```python
percentiles = {
	"0_1": int(ROWS_COUNT * 0.001),
	"1": int(ROWS_COUNT * 0.01),
	"3": int(ROWS_COUNT * 0.03),
	"10": int(ROWS_COUNT * 0.1),
}

for p, row_count in percentiles.items():
	with open("./data/1m_6ex_karel/train.json", "r") as f:
		with open(f"./data/1m_6ex_karel/train_p_{p}.json", "w") as f_p:
			for _ in range(row_count):
				f_p.write(f.readline())
```

```python
import json

def get_dataset_row_iter(filename: str):
	with open(filename, "r") as f:
		for line in f:
			yield json.loads(line)

it = get_dataset_row_iter(f"./data/1m_6ex_karel/train.json")
print(next(it))
```

```python
import os
import json
import psutil
import gc
import multiprocessing as mp

def rss_mb():
    return psutil.Process(os.getpid()).memory_info().rss / (1024 ** 2)

def iterate_worker(path, q):
	gc.collect()
	start = rss_mb()
	peak = start
	cnt = 0

	it = get_dataset_row_iter(path)
	for line in it:
		row = line
		cnt += 1
			
		if cnt % 1000 == 0:
			peak = max(peak, rss_mb())

	end = rss_mb()
	q.put({
		"rows": cnt,
		"start_mb": start,
		"end_mb": end,
		"delta_mb": end - start,
		"peak_mb": max(peak, end)
	})

def run_in_subprocess(worker, path):
    q = mp.Queue()
    p = mp.Process(target=worker, args=(str(path), q))
    p.start()
    p.join()
    return q.get()

iter_res = run_in_subprocess(iterate_worker, "./data/1m_6ex_karel/train.json")

print(iter_res)
```
